As I have searched online, I found out the peak traffic in Addis Ababa city is in morining ang evening and roughly lasts for 2-3 hours, so it is advisible to divide Time column in buckets of 3 hours starting from 0800 hours.

```python
rta_data["Time"] = pd.to_datetime(rta_data["Time"])
```

```python
def divide_time(time_):
    time_ = time_.to_pydatetime().time()
    first = time(8, 0, 0)
    second = time(11, 0, 0)
    third = time(14, 0, 0)
    forth = time(17, 0, 0)
    fifth = time(20, 0, 0)
    sixth = time(23, 0, 0)
    seventh = time(2, 0, 0)
    eighth = time(5, 0, 0)
    if time_ >= first and time_ < second:
        return "first"
    elif time_ >= second and time_ < third:
        return "second"
    elif time_ >= third and time_ < forth:
        return "third"
    elif time_ >= forth and time_ < fifth:
        return "forth"
    elif time_ >= fifth and time_ < sixth:
        return "fifth"
    elif time_ >= sixth and time_ < seventh:
        return "sixth"
    elif time_ >= seventh and time_ < eighth:
        return "seventh"
    else: 
        return "eighth"     
```

```python
rta_data.Time = rta_data.Time.apply(divide_time)
```
